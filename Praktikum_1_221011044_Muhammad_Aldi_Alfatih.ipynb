Hasilnya adalah gambar baru yang menampilkan hanya bagian yang berwarna kuning, sementara bagian lainnya hilang. Ini berguna untuk fokus pada warna tertentu dalam gambar tanpa gangguan dari elemen lain.

```python
plt.figure(1)
plt.subplot(121)
plt.imshow(tulips[:,:,::-1])  # Gambar asli
plt.title("Gambar Asli")

plt.subplot(122)
plt.imshow(res[:,:,::-1])  # Gambar hasil ekstraksi warna kuning
plt.title("Ekstraksi Kuning")
```
